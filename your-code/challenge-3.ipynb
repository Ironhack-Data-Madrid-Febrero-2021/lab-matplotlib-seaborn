#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way
plt.scatter(titanic.Fare,titanic.Age)
plt.title('Scatter plot between Titanic´s Fare and passenger´s Age')
plt.xlabel('Titanic Fare')
plt.ylabel('Passenger´s Age')
```

```python
# your code-2nd way
correlation_coefficients = titanic.corr()
correlation_coefficients
```

```python
sns.regplot(x=titanic.Age, y=titanic.Fare);
```

```python
#As seen in the two methods above, the behavior of the variable
#Fare is explained in a 17% by the variable age. 
#Therefore we could conclude that these two variables are not correlated.
```
